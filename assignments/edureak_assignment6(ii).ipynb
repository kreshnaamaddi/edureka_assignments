```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (replace 'hurricanes.csv' with your actual filename)
df = pd.read_csv('hurricanes.csv')

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(df['Year'], df['Hurricanes'], color='skyblue')
plt.xlabel('Year')
plt.ylabel('Number of Hurricanes')
plt.title('Number of Hurricanes Along the Atlantic Coast (US)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Load the dataset (replace 'city_temperature.csv' with actual filename)
df = pd.read_csv('city_temperature.csv')

# Filter data for San Francisco and Moscow
sf_temps = df[df['City'] == 'San Francisco']['Temperature']
moscow_temps = df[df['City'] == 'Moscow']['Temperature']

# Plot histograms
plt.figure(figsize=(10, 6))
plt.hist(sf_temps.dropna(), bins=20, alpha=0.7, label='San Francisco', color='orange')
plt.hist(moscow_temps.dropna(), bins=20, alpha=0.7, label='Moscow', color='blue')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.title('Temperature Distribution (2014â€“2015)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Load the dataset (replace 'car_models.csv' with actual filename)
df = pd.read_csv('car_models.csv')

# Count the number of models by manufacturer
model_counts = df['Manufacturer'].value_counts()

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(model_counts, labels=model_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Number of Models Released by Each Manufacturer')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()

# Find the manufacturer with the most releases
top_manufacturer = model_counts.idxmax()
print("Manufacturer with the most releases:", top_manufacturer)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

X = [1, 2, 3, 4]
y = [20, 21, 20.5, 20.8]

plt.plot(X, y)
plt.show()
```

```python
plt.plot(X, y, color='green', linestyle='--', marker='o', linewidth=2, markersize=8)
plt.show()
```

```python
plt.plot(X, y)
plt.axis([0, 5, 19, 22])  # x: 0 to 5, y: 19 to 22
plt.show()
```

```python
plt.plot(X, y, marker='o')
plt.title('Temperature Variation', fontsize=14)
plt.xlabel('Day', fontsize=14)
plt.ylabel('Temperature (Â°C)', fontsize=14)
plt.grid(True)
plt.show()
```

```python
y_error = [0.12, 0.13, 0.2, 0.1]
plt.errorbar(X, y, yerr=y_error, fmt='-o', color='purple', ecolor='red', capsize=5)
plt.title('Temperature with Error Bars', fontsize=14)
plt.xlabel('Day', fontsize=14)
plt.ylabel('Temperature (Â°C)', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(4, 5), dpi=100)
plt.plot(X, y, marker='o')
plt.title('Fixed Size & DPI', fontsize=14)
plt.xlabel('Day', fontsize=14)
plt.ylabel('Temperature (Â°C)', fontsize=14)
plt.tight_layout()
plt.show()
```
