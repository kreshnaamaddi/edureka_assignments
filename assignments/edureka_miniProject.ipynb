```python
# Simple Caesar Cipher Encryption/Decryption with shift = 3
def encrypt(password):
    return ''.join(chr(ord(c) + 3) for c in password)

def decrypt(password):
    return ''.join(chr(ord(c) - 3) for c in password)

# File path to store credentials
FILE_NAME = "credentials.txt"

def add_credential():
    website = input("Website: ")
    username = input("Username: ")
    password = input("Password: ")
    encrypted_password = encrypt(password)
    with open(FILE_NAME, "a") as file:
        file.write(f"{website}|{username}|{encrypted_password}\n")
    print("Credential saved successfully.\n")

def view_all_credentials():
    try:
        with open(FILE_NAME, "r") as file:
            lines = file.readlines()
            if not lines:
                print("No credentials stored.\n")
                return
            print("Stored Credentials:")
            for line in lines:
                website, username, encrypted_password = line.strip().split("|")
                print(f"Website: {website}, Username: {username}, Password: {encrypted_password}")
    except FileNotFoundError:
        print("No data file found. Start adding credentials.\n")

def search_by_website():
    site = input("Enter website to search: ").strip().lower()
    found = False
    try:
        with open(FILE_NAME, "r") as file:
            for line in file:
                website, username, encrypted_password = line.strip().split("|")
                if website.lower() == site:
                    print("Found:")
                    print(f"Website: {website}")
                    print(f"Username: {username}")
                    print(f"Password: {encrypted_password}")
                    found = True
                    break
        if not found:
            print("No credentials found for the given website.\n")
    except FileNotFoundError:
        print("No data file found. Start adding credentials.\n")

def main():
    while True:
        print("--- Password Manager ---")
        print("1. Add Credential")
        print("2. View All")
        print("3. Search by Website")
        print("4. Exit")
        choice = input("Select an option: ")

        if choice == "1":
            add_credential()
        elif choice == "2":
            view_all_credentials()
        elif choice == "3":
            search_by_website()
        elif choice == "4":
            print("Exiting Password Manager.")
            break
        else:
            print("Invalid choice. Please try again.\n")

if __name__ == "__main__":
    main()
```
